__1.__ (2 pts) Write the definitions of convex and strongly convex functions. Provide the example of strongly convex function and find the corresponding constant of strongly convexity

Answer:

Function $f : X \subset \mathbb{R^n} \to \mathbb{R}$ is convex if $X$ is a convex set and $∀x_1, x_2 \in X, a \in [0,1]$ the following holds:
$$f(ax_1 + (1-a)x_2) \leq af(x_1) + (1-a)f(x_2)$$

Function $f : X \subset \mathbb{R^n} \to \mathbb{R}$ is strongly convex with constant $m > 0$ if $X$ is a convex set and $∀x_1, x_2 \in X, a \in [0,1]$ the following holds:
$$f(ax_1 + (1-a)x_2) \leq af(x_1) + (1-a)f(x_2) - \frac{m}{2}a(1-a)\|x_1 - x_2\|_2^2$$

The set $X$ is convex if $∀x_1,x_2 \in X, ∀a \in [0,1]$ the following holds:
$$ax_1 + (1-a)x_2 \in X$$

Example:

$f(x) = x^2$ is a strongly convex function. Moreover, it is differentiable.

Therefore, an equivalent definition for strongly convex function is:

$f(x_1) ≥ f(x_2) + f'(x_2)(x_1-x_2) + \frac{m}{2}\|x_1-x_2\|_2^2$

$f'(x) = 2x$

Therefore:
$x_1^2 ≥ x_2^2 + 2x_2(x_1-x_2) + \frac{m}{2}\|x_1-x_2\|_2^2$

Since $\|x_1-x_2\|_2^2 = {\sqrt{(x_1 - x_2)^2}}^2 = (x_1-x_2)^2$, we have:

$x_1^2 ≥ x_2^2 + 2x_2(x_1-x_2) + \frac{m}{2}(x_1-x_2)^2$

Hence:

$x_1^2 ≥ x_2^2 - 2x_2^2 + 2x_1x_2 + \frac{m}{2}(x_1^2 -2x_1x_2 + x_2^2)$

It can be seen that if $m=2$, the whole inequality becomes the following:

$x_1^2 \ge x_1^2$, hence $x_1^2 = x_1^2$

Finally, that means, we can choose $m=2$ as the constant for strong convexity




__2.__ (2 pts) Write the definition of convex set. Check if the following set $G$ is convex or not and explain your result.

$$
G = \{ x \in \mathbb{R}^n \mid A_0 + x_1A_1 + \ldots + x_nA_n \succeq 0 \},
$$

where all $A_i$ are symmetric matrices $n \times n$, the notation $X \succeq 0$ means that matrix $X$ is symmetric positive semidefinite.

The set $X ⊆ \mathbb{R^n}$ is convex if $∀x_1,x_2 \in X, ∀a \in [0,1]$ the following holds:
$$ax_1 + (1-a)x_2 \in X$$

Let $x,y \in G$ and let $a \in [0,1]$

Let $z = ax + (1-a)y$

We need to show that $z \in G$. Hence we must show that $A_0 + z_1A_1 + \ldots + z_nA_n \succeq 0$

We start as follows:

$$A_0 + z_1A_1 + \ldots + z_nA_n = A_0 + (ax_1+(1-a)y_1)A_1 + \ldots + (ax_n+(1-a)y_n)A_n \\ = A_0 + ax_1A_1 + (1-a)y_1A_1 + \ldots + ax_nA_n + (1-a)y_nA_n \\ = A_0 + a(x_1A_1 + \ldots x_nA_n) + (1-a)(y_1A_1 + \ldots y_nA_n)$$

We can note that $A_0 = aA_0 + (1-a)A_0$, since $aA_0 + (1-a)A_0 = aA_0 + A_0 - aA_0 = A_0$

Hence,

$$A_0 + a(x_1A_1 + \ldots x_nA_n) + (1-a)(y_1A_1 + \ldots y_nA_n) \\ = aA_0 + (1-a)A_0 + a(x_1A_1 + \ldots x_nA_n) + (1-a)(y_1A_1 + \ldots y_nA_n) \\ = a(A_0 + x_1A_1 + \ldots x_nA_n) + (1-a)(A_0 + y_1A_1 + \ldots y_nA_n)$$

We know that $A_0 + x_1A_1 + \ldots x_nA_n \succeq 0$ and $A_0 + y_1A_1 + \ldots y_nA_n \succeq 0$, since $x,y \in G$

We also know that $0 ≤ a ≤ 1$, hence $a(A_0 + x_1A_1 + \ldots x_nA_n) + (1-a)(A_0 + y_1A_1 + \ldots y_nA_n) \succeq 0$ holds, since as sum of 2 positive semi-definite matrices is also a positive semi-definite matrix

Finally, that means:

$A_0 + z_1A_1 + \ldots + z_nA_n \succeq 0$ from above, hence $z \in G \\ \blacksquare$

__3.__ Consider the following problem

$$
\min_w \frac{1}{2}\|w\|_2^2 + \sum_{i=1}^m \max(0, 1 - y_iw^{\top}x_i),

- (1 pts) Is this problem convex or not? Explain your answer

- (2 pts) Write the KKT conditions for smooth form of this problem or to the original problem if you decide that it is smooth already
- (1 pts) Are KKT conditions sufficient for this problem? Why? 

1. Max function is not differentiable, hence the problem is not smooth
2. $\|w\|_2^2$ is convex by definition. Scaling a convex function by a positive constant preserves its convexity. Hence $\frac{1}{2}\|w\|_2^2$ is convex as well.

  $\max(0, 1 - y_iw^{\top}x_i)$ is convex by definition of a maximum function. Sum of $m$ convex functions gives a convex function as well, since the sum presesrves the convexity. So $\sum_{i=1}^m \max(0, 1 - y_iw^{\top}x_i)$ is also convex.

  Lastly the whole expression is also convex because it represents again the sum of 2 convex functions.

3. $\sum_{i=1}^m \max(0, 1 - y_iw^{\top}x_i)$ can be transformed to $\sum_{i=1}^m 1 - y_iw^{\top}x_i$ with a constraint that $1 - y_iw^{\top}x_i \ge 0$

  We can simplify this even further: $\sum_{i=1}^m 1 - y_iw^{\top}x_i$ is equivalent to $m - \sum_{i=1}^m y_iw^{\top}x_i$

  Hence, the final problem would be:

  $$
    \min_w \frac{1}{2}\|w\|_2^2 + m - \sum_{i=1}^m y_iw^{\top}x_i
  $$

  $$s.t. y_iw^{\top}x_i ≤ 1 \  ∀i \in \{1, \ldots, m\}$$

4. Before we find KKT conditions, we need to define Lagrangian and find its derivative with respect to parameter $w$
$$L(w,\lambda, μ) = \frac{1}{2}\|w\|_2^2 + m - \sum_{i=1}^m y_iw^⊤x_i + \sum_{i=1}^m \mu_i(y_iw^⊤x_i - 1)$$

$$L'_w(w,\lambda, μ) = w - \sum_{i=1}^m y_ix_i + \sum_{i=1}^m \mu_iy_ix_i$$

  KKT conditions for the smooth form of the problem:

$$
1) y_iw^⊤x_i - 1 ≤ 0, \ ∀i \in \{1, \ldots, m\} \\
2) μ_i ≥ 0, \ ∀i \in \{1, \ldots, m\} \\
3) μ_i(y_iw^⊤x_i - 1) = 0, \ ∀i \in \{1, \ldots, m\} \\
4) L'_w(w,\lambda, μ) = w - \sum_{i=1}^m y_ix_i + \sum_{i=1}^m \mu_iy_ix_i = 0
$$

  5. Since the problem is convex, KKT conditions are necessary and sufficient

```python
import jax
import jax.numpy as jnp

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import liboptpy.unconstr_solvers.fo as fo
import liboptpy.step_size as ss
import liboptpy.base_optimizer as base

import cvxpy as cvx
print(cvx.installed_solvers())
```

```python
def Newton(f, gradf, hessf, x0, alpha, epsilon = 1e-8, num_iter = 50):
    x = x0
    iteration = 0
    
    norms = []

    xs = [x0]

    while True:
        gradient = gradf(x)
        hessian = hessf(x)

        # From FOOC since function F is convex
        happrox = -gradient / hessian

        x = x + alpha * happrox

        xs.append(x)

        iteration += 1
        # print("Current function val =", f(x))
        # print("Current gradient norm = ", jnp.linalg.norm(gradf(x)))

        norms.append(jnp.linalg.norm(gradf(x)))

        if jnp.linalg.norm(gradf(x)) < epsilon:
            break
        if iteration >= num_iter:
            break
    res = {"x": x, "xs": xs, "num_iter": iteration, "tol": jnp.linalg.norm(gradf(x)), "norms": norms}
    return res
```

```python
def F(x):
  return jnp.log(jnp.exp(x) + jnp.exp(-x))
```

```python
gradf = jax.grad(F)
hessf = jax.hessian(F)
```

```python
def PlotResults(res, title, methodName):
  plt.figure(figsize=(10,6))
  plt.title(title)
  plt.plot(jnp.arange(1, len(res) + 1), res, label=methodName)
  plt.xlabel(r"Number of iterations, $k$", fontsize=15)
  plt.ylabel("Gradient norm", fontsize=15)
  # plt.yscale("log")
  plt.legend(loc="best", fontsize=15)

def PlotFunction():
  x = x = jnp.linspace(-1000,1000,10000)
  y = [F(x_i) for x_i in x]

  plt.figure(figsize=(10,6))
  plt.plot(x, y, label="Function")
  plt.xlabel(r"$x$", fontsize=15)
  plt.ylabel(r"f(x)", fontsize=15)
  plt.legend(loc="best", fontsize=15)

def PlotFunctionWithResults(res, title, methodName):
  x = x = jnp.linspace(-4,4,10000)
  y = [F(x_i) for x_i in x]

  yres = [F(res_x) for res_x in res]

  plt.figure(figsize=(10,6))
  plt.title(title)
  plt.plot(x, y, label="Function")
  plt.plot(res, yres, alpha=0.7, color='r', lw=3, label=methodName)
  plt.xlabel(r"$x$", fontsize=15)
  plt.ylabel(r"f(x)", fontsize=15)
  plt.legend(loc="best", fontsize=15)
```

```python
x0 = 1.1
res2 = Newton(F, gradf, hessf, x0, alpha = 1, epsilon = 1e-8, num_iter = 50)
PlotResults(res2['norms'], "Convergence plot with x0 = 1.1", "Newton method")
PlotFunctionWithResults(res2['xs'], "Newton solution with initial x = 1.1", "Newton method")
```

So we can see that this method is very sensitive to the initial point x0

Step size for steepest descent:

$$f(x) = \frac12 \|Ax - b\|_2^2$$

$$\nabla f(x) = A^\top (Ax - b)$$

$$\nabla^2 f(x) = A^\top A$$

$x_{k+1} = x_k - α_k\nabla f(x_k)$

Since we are using steepest descent rule, we know that (1) $\nabla f(x_{k+1})^⊤ \nabla f(x_k) = 0$, That is, at every iteration, the gradient steps are orthogonal to each other.

$$\nabla f(x_{k+1}) = A^⊤ (A x_{k+1} - b) = A^⊤ (A(x_k - a_k \nabla f(x_k)) - b) \\ = A^⊤ (Ax_k - a_kA \nabla f(x_k) - b) = A^⊤ (Ax_k - b - a_kA \nabla f(x_k)) \\ = \nabla f(x_k) - a_k A^⊤ A \nabla f(x_k)$$


$$\nabla f(x_{k+1})^⊤ = \nabla f(x_k)^⊤ - a_k A^⊤ A \nabla f(x_k)^⊤$$

Therefore from (1):

$$(\nabla f(x_k)^⊤ - a_k A^⊤ A \nabla f(x_k)^⊤) \nabla f(x_k) = 0$$
$$\nabla f(x_k)^⊤ \nabla f(x_k) - a_k \nabla f(x_k)^⊤ A^⊤ A \nabla f(x_k) = 0$$
$$\nabla f(x_k)^⊤ \nabla f(x_k) = a_k \nabla f(x_k)^⊤ A^⊤ A \nabla f(x_k)$$
$$a_k = \frac{\nabla f(x_k)^⊤ \nabla f(x_k)}{\nabla f(x_k)^⊤ A^⊤ A \nabla f(x_k)}$$

```python
A = np.random.rand(1000,100)
x0 = np.random.rand(100)
b = np.random.rand(1000)

epsilon = 10e-4
numIter = 10000

def fx(x):
  return 0.5 * np.linalg.norm(A @ x - b) ** 2

def gradfx(x):
  return A.T @ (A @ x - b)

def hessfx(x):
  return A.T @ A
```

```python
def SteepestGradientDescent(f, gradf, x0, epsilon, num_iter):
    x = x0.copy()
    iteration = 0

    norms = list()

    while True:
        gradient = gradf(x)

        alpha = (gradient.T @ gradient) / (gradient.T @ A.T @ A @ gradient)
        
        x -= alpha * gradient

        iteration += 1

        norm = np.linalg.norm(gradf(x))

        norms.append(norm)

        if np.linalg.norm(gradf(x)) < epsilon:
            break
        if iteration >= num_iter:
            break
    res = {"x": x, "num_iter": iteration, "tol": np.linalg.norm(gradf(x)), "norms": norms}
    return res
```

```python
class HeavyBall(base.LineSearchOptimizer):
    def __init__(self, f, grad, step_size, beta, **kwargs):
        super().__init__(f, grad, step_size, **kwargs)
        self._beta = beta
    
    def get_direction(self, x):
        self._current_grad = self._grad(x)
        return -self._current_grad

    def _f_update_x_next(self, x, alpha, h):
        if len(self.convergence) < 2:
            return x + alpha * h
        else:
            return x + alpha * h + self._beta * (x - self.convergence[-2])
    
    def get_stepsize(self):
        return self._step_size.get_stepsize(self._grad_mem[-1], self.convergence[-1], len(self.convergence))
```

```python
hessian = hessfx(x0)

maxEigHessian = max(np.linalg.eig(hessian)[0])
minEigHessian = min(np.linalg.eig(hessian)[0])

alpha_opt_hb = 4 / (np.sqrt(maxEigHessian) + np.sqrt(minEigHessian))**2 
beta_opt_hb = np.maximum((1 - np.sqrt(alpha_opt_hb * maxEigHessian))**2, 
                      (1 - np.sqrt(alpha_opt_hb * minEigHessian))**2)

optStepSize = 1 / maxEigHessian

print(f"Using optimal step size = {optStepSize}")

print(f"alpha_opt_hb = {alpha_opt_hb}, beta_opt_hb = {beta_opt_hb}")
```

```python
methods = {
    "GD Optimal": fo.GradientDescent(fx, gradfx, ss.ConstantStepSize(optStepSize)),
    "GD Armijo": fo.GradientDescent(fx, gradfx, ss.Backtracking("Armijo", rho=0.5, beta=0.1, init_alpha=0.5)),
    "GD Nesterov": fo.AcceleratedGD(fx, gradfx, ss.ConstantStepSize(optStepSize)),
    "Heavy Ball": HeavyBall(fx, gradfx, ss.ConstantStepSize(alpha_opt_hb), beta=beta_opt_hb)
}
```

```python
for m in methods:
    _ = methods[m].solve(x0=x0, max_iter=numIter, tol=epsilon)

res4 = SteepestGradientDescent(fx, gradfx, x0, epsilon, numIter)
```

```python
plt.rc("text", usetex=False)

figsize = (8, 6)
fontsize = 14
plt.figure(figsize=figsize)
for m in methods:
    plt.semilogy([np.linalg.norm(gradfx(x)) for x in methods[m].get_convergence()], label=m, linewidth=3)

plt.semilogy(res4['norms'], label="Steepest GD", linewidth=3)

plt.legend(fontsize=fontsize)
plt.xlabel("Number of iteration, $k$", fontsize=fontsize)
plt.ylabel(r"$\| f'(x_k)\|_2$", fontsize=fontsize)
plt.xticks(fontsize=fontsize)
_ = plt.yticks(fontsize=fontsize)
```

Out of the three strategies: steepest descent rule, optimal constant step size and Armijo rule, the least iterations to converge took the Steepest GD method, then the GD with Armijo rule and most iterations took GD with optimal step size.

Comparing Heavy ball and Nesterov GD, they converge faster than steepest GD, especially Heavy Ball method, it converges the fastest.
